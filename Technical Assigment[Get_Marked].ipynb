```python
from typing import List, Tuple
import random
```

```python
def snapshot(num_people: int = 20_000, hall_length: int = 250, hall_breath: int = 300) -> List[Tuple[str, float, float]]:
    """ There's 20,000 people in a conference hall of 250m by 300m

    Returns:
        a list of 3-tuple of Person_id, x-coordinate, y-coordinate

    """
    return [(f'Person{i:06}', random.random() * hall_length, random.random() * hall_breath,) for i in range(num_people)]
```

```python
#Algorithm

#Set up a first peopleID_with_coordinates dictionary with PersonID: (x_coordintate, y_coordinate)

#Set up a second dictonary peopleID_with_violations with data structure of  key[key[value]] 
#Key Person ID. 
#2nd Key is the Interacting Person Id
#Value: 1 if within 2 meter radius, 0 if not

#Compare first person in totalNoOfPeople with the next n-1 people
    #For each person, if the radius (defined by the euclidean distance between x and y coordinates) <=2,
        #Mark the Interacting Person Id = 1
#return the dictionary peopleID_with_violations
```

```python
def social_distancing_checker(people_in_conferenceHall):
    peopleID_with_coordinates = {people_in_conferenceHall[ele][0]:(people_in_conferenceHall[ele][1],
                                                                   people_in_conferenceHall[ele][2])
                                for ele in range(len(people_in_conferenceHall))}
    peopleID_with_violations = {}
    checker = []
    for key, value in peopleID_with_coordinates.items():
        curr_Person = key
        curr_x_coordinate = value[0] #Index 0- x coordinates, Index 1- y coordinates
        curr_y_coordinate =  value[1]
        checker.append(curr_Person)
        peopleID_with_violations[curr_Person] = {}
        for key, value in peopleID_with_coordinates.items():
            next_person = key 
            next_x_coordinate = value[0]
            next_y_coordinate = value[1]
            if key == curr_Person: continue
            if ((next_x_coordinate-curr_x_coordinate)**2 + (next_y_coordinate-curr_y_coordinate)**2) <=2 :
                peopleID_with_violations[curr_Person][next_person] = 1
    return peopleID_with_violations
```

```python
people_in_conferenceHall = snapshot() 
```

```python
peopleID_with_coordinates = {people_in_conferenceHall[ele][0]:(people_in_conferenceHall[ele][1],people_in_conferenceHall[ele][2])
                                for ele in range(len(people_in_conferenceHall))}
social_distance_dictionary = social_distancing_checker(people_in_conferenceHall)
```
